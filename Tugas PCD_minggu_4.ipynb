##**1. Konversi RGB ke Grayscale**
Keterangan:
Konversi RGB ke grayscale dilakukan dengan menggabungkan tiga channel warna (Red, Green, Blue) menjadi satu nilai intensitas.

```python
import cv2
import matplotlib.pyplot as plt

img = cv2.imread('/content/patrick.jpg')
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

plt.imshow(gray, cmap='gray')
plt.title('Grayscale Image')
plt.axis('off')
plt.show()
```

##**2. Konversi Grayscale ke Biner**

Keterangan:
Proses ini disebut thresholding, yaitu mengubah citra grayscale menjadi dua nilai saja:
0 (hitam)
255 (putih)
Jika pixel > threshold (127), maka menjadi putih, jika tidak maka hitam.

```python
import cv2
import matplotlib.pyplot as plt

_, binary = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY)

plt.imshow(binary, cmap='gray')
plt.title('Binary Image')
plt.axis('off')
plt.show()
```

##**3. Konversi Grayscale ke m-bit**

Keterangan:
Kuantisasi mengurangi jumlah level keabuan:
8 bit → 256 level
4 bit → 16 level

Semakin kecil bit, maka detail gambar yang terlihat akan semakin berkurang

```python
import cv2
import matplotlib.pyplot as plt
mbit = (gray // 16) * 16

plt.imshow(mbit, cmap='gray')
plt.title('4-bit Image (16 Level)')
plt.axis('off')
plt.show()
```

##**4. image dengan brightness**

Keterangan:
Brightness digunakan untuk mengatur kecerahan

```python
import cv2
import matplotlib.pyplot as plt
brightness = cv2.convertScaleAbs(gray, alpha=1, beta=50)

plt.imshow(brightness, cmap='gray')
plt.title('Brightness Increased')
plt.axis('off')
plt.show()
```

##**5. Image dengan contrast**

Keterangan:
Digunakan untuk mengatur perbedaan intensitas antar pixel

```python
import cv2
import matplotlib.pyplot as plt
contrast = cv2.convertScaleAbs(gray, alpha=1.5, beta=0)

plt.imshow(contrast, cmap='gray')
plt.title('Contrast Increased')
plt.axis('off')
plt.show()
```

##**6. Histogram Grayscale**
Keterangan:
Histogram menunjukkan distribusi intensitas pixel.
kiri → gelap
kanan → terang
Digunakan untuk analisis kualitas citra.

```python
import cv2
import matplotlib.pyplot as plt
plt.hist(gray.ravel(), bins=256, range=[0,256])
plt.title('Histogram Grayscale')
plt.show()
```

##**7. Histogram Equalization**

Keterangan:
Histogram equalization digunakan untuk meningkatkan kontras dengan meratakan distribusi intensitas.
Tujuan:
1. memperjelas detail
2. memperbaiki citra yang terlalu gelap/terang

```python

equalized = cv2.equalizeHist(gray)

plt.imshow(equalized, cmap='gray')
plt.title('Histogram Equalization')
plt.axis('off')
plt.show()
```
